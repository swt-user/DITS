```python
def parse_line(line):
    # 分割出索引和数据部分
    parts = line.split(':', 1)  # 最多分割2次
    index = int(parts[0])
    # print(parts)
    # 处理数据部分
    data_parts = parts[1].split(';')
    values = {}
    for part in data_parts:
        key_value = part.strip().split(':')
        key = key_value[0].strip()
        value = float(key_value[1].strip())
        values[key] = value
    return int(index), values
    
# 找出同时在前50%的数据
def find_top_50_percent(data, percent=0.7):
    # 分别获取每个属性的值列表
    margins = [d['margins'] for d in data.values()]
    accuracies = [d['accuracies'] for d in data.values()]
    losses = [d['loss'] for d in data.values()]
    
    # 计算每个属性的中位数
    margin_median = sorted(margins)[int(0.5*len(margins))]
    accuracy_median = sorted(accuracies)[int(0.5*len(accuracies))]
    loss_median = sorted(losses)[int(0.5*len(losses))]
    
    # 找出同时满足条件的数据
    result = {}
    for idx, values in data.items():
        # if (values['margins'] >= margin_median and 
        #     values['accuracies'] >= accuracy_median and 
        #     values['loss'] <= loss_median):  # 注意loss是越小越好
        #     result[idx] = values
        if values['loss'] <= loss_median and values['margins'] >= margin_median and values['accuracies'] >= accuracy_median:
            result[idx] = values
    return result

# 示例使用
result = {}
for start_point in [0, 200,400,600]:
    with open(f"/home/wentaos/Optima/local_influence/case0_1example_1559_{start_point}.txt", 'r') as f:
        for line in f:
            index, value = parse_line(line)
            result[index] = value
new_result = find_top_50_percent(result)
selected_chosen_value = []
selected_distance = []

for key, value in new_result.items():
    # print(str(value)+'\n'+f"chosen_value:{chosen_value[key]}; distance:{distance[key]}")
    selected_chosen_value.append(chosen_value[key])
    selected_distance.append(distance[key])
print(len(selected_chosen_value))
print(mean(selected_chosen_value))
print(mean(selected_distance))
```

```python
import pandas as pd
from scipy.stats import pearsonr

# 创建一个空的 DataFrame 来存储数据
data = {
    'id': [],
    'best_1000': [],
    'best_100': [],
    'count_1000': [],
    'count_100': []
}
i = 0
while i < len(lines):
    id_ = lines[i].split('/')[-1]
    i += 1
    best_1000 = float(lines[i].split(': ')[1])
    i += 1
    count_1000 = int(lines[i].split(': ')[1])
    i += 1
    best_100 = float(lines[i].split(': ')[1])
    i += 1
    count_100 = int(lines[i].split(': ')[1])
    i += 1
    
    data['id'].append(id_)
    data['best_1000'].append(best_1000)
    data['best_100'].append(best_100)
    data['count_1000'].append(count_1000)
    data['count_100'].append(count_100)

# 创建 DataFrame
df = pd.DataFrame(data)

# 计算 best_1000 和 best_100 的相关性
correlation, p_value = pearsonr(df['best_1000'], df['best_100'])

print(f"Pearson correlation between best_1000 and best_100: {correlation}")
print(f"P-value: {p_value}")

# 输出 DataFrame 以检查数据
print(df.head())
```
